```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
```

```python
# Standardize the features using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(dt)

# Find the optimal eps value using the k-distance graph
neigh = NearestNeighbors(n_neighbors=2)
nbrs = neigh.fit(X_scaled)
distances, indices = nbrs.kneighbors(X_scaled)
```

```python
# Sort the distances and plot the k-distance graph
distances = np.sort(distances, axis=0)
distances = distances[:, 1]
plt.plot(distances)
plt.xlabel('Data Points')
plt.ylabel('Epsilon (k-distance)')
plt.title('K-Distance Graph')
plt.show()
```

```python
optimal_eps_value = 1.5

# Apply DBSCAN with the optimal eps value
dbscan = DBSCAN(eps=optimal_eps_value, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Add the DBSCAN labels to the original dataset
dt['dbscan_labels'] = labels

# Remove unclustered data (Cluster id = 0)
filtered_dt = dt[dt['dbscan_labels'] != 0]
```

```python
dt= filtered_dt
```

```python
#XGBoost

import xgboost as xgb
xgb = xgb.XGBClassifier(n_estimators= 800)
xgb.fit(X_train,y_train)
y_pred_xgb = xgb.predict(X_test)

CM=confusion_matrix(y_test,y_pred_xgb)
sns.heatmap(CM, annot=True)
acc= accuracy_score(y_test, y_pred_xgb)
print(f'ACCURACY : {acc}')
```

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as pl

k_values = list(range(3, 21))  # Adjust the range as needed

# Create a KFold cross-validation object
kf = KFold(n_splits=5, shuffle=True, random_state=42)  # Adjust the number of splits as needed

# Lists to store mean cross-validation scores for each k
cv_scores_mean = []

# Loop over each k value
for k in k_values:
    # Create a KNN classifier with the current k value
    knn = KNeighborsClassifier(n_neighbors=k)

    # Perform cross-validation and get the mean accuracy score
    scores = cross_val_score(knn, X, y, cv=kf, scoring='accuracy')
    cv_scores_mean.append(np.mean(scores))

# Plot the results to visualize the performance for different k values
plt.plot(k_values, cv_scores_mean, marker='o')
plt.title('Cross-Validation Scores for Different k Values')
plt.xlabel('k')
plt.ylabel('Mean Accuracy')
plt.show()

# Find the k value with the highest cross-validation score
best_k = k_values[np.argmax(cv_scores_mean)]
print(f'The optimal k value is: {best_k}')
```

```python
#K nearest neighbour (Cross-Validation for finding n value)

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(best_k)
knn.fit(X_train,y_train)
y_pred_knn = knn.predict(X_test)


CM=confusion_matrix(y_test,y_pred_knn)
sns.heatmap(CM, annot=True)
acc= accuracy_score(y_test, y_pred_knn)
print(f'ACCURACY : {acc}')
```

```python
#Support Vector Classifier (kernel='linear')
from sklearn.svm import SVC
svc = SVC(kernel='linear',gamma='auto',probability=True)
svc.fit(X_train,y_train)
y_pred_svc = svc.predict(X_test)


CM=confusion_matrix(y_test,y_pred_svc)
sns.heatmap(CM, annot=True)
acc= accuracy_score(y_test, y_pred_svc)
print(f'ACCURACY : {acc}')
```

```python
#Random Forest Classifier (criterion = 'entropy')

rf_ent = RandomForestClassifier(criterion='entropy',n_estimators=200)
rf_ent.fit(X_train, y_train)
y_pred_rfe = rf_ent.predict(X_test)

CM=confusion_matrix(y_test,y_pred_rfe)
sns.heatmap(CM, annot=True)
acc= accuracy_score(y_test, y_pred_rfe)
print(f'ACCURACY : {acc}')
```
